```python
farm_data = farm_data[farm_data["Broilers and other meat-type chickens"].str.contains("(D)")==False]
farm_data = farm_data[farm_data["Cattle and calves"].str.contains("(D)")==False]
farm_data = farm_data[farm_data["Hogs and pigs"].str.contains("(D)")==False]
farm_data = farm_data[farm_data["Sheep and lambs"].str.contains("(D)")==False]
farm_data = farm_data[farm_data["Land in Farms"].str.contains("(D)")==False]
farm_data = farm_data[farm_data["Pastureland"].str.contains("(D)")==False]
```

```python
farm_data2 = farm_data 
farm_data2
```

```python
farm_data.replace(",",'', inplace=True, regex=True)
farm_data
```

```python
farm_data2
state_county = farm_data2[['State','County']]
```

Pastureland * Land in Farms = Acres of Pastureland

```python
cols = farm_data2.columns[2:]
cols


farm_data_num = farm_data2[cols].apply(pd.to_numeric, errors = 'coerce')
farm_data_num.info()
farm_data_num
```

```python
cleaned_farm_data = pd.concat([state_county, farm_data_num], axis=1)
cleaned_farm_data
```

```python
cleaned_farm_data["Pastureland Acres"] = cleaned_farm_data["Pastureland"]*cleaned_farm_data["Land in Farms"]/100
cleaned_farm_data
```

```python
florida = cleaned_farm_data.loc[farm_data['State']=='Florida']
```

```python
print("Number of pigs: " + str(florida['Hogs and pigs'].sum()))
print("Number of cows: " + str(florida['Cattle and calves'].sum()))
print("Number of sheep: " + str(florida['Sheep and lambs'].sum()))
print("Number of chickens: " + str(florida['Broilers and other meat-type chickens'].sum()))
```

```python
cleaned_farm_data["Chicken CH4 per Acre (Kg/Year)"] = cleaned_farm_data['Broilers and other meat-type chickens']*.1/cleaned_farm_data["Pastureland Acres"]
cleaned_farm_data["Cow CH4 per Acre (Kg/Year)"] = cleaned_farm_data['Cattle and calves']*69.5/cleaned_farm_data["Pastureland Acres"]
cleaned_farm_data["Sheep CH4 per Acre (Kg/Year)"] = cleaned_farm_data['Sheep and lambs']*.5/cleaned_farm_data["Pastureland Acres"]
cleaned_farm_data["Pig CH4 per Acre (Kg/Year)"] = cleaned_farm_data['Hogs and pigs']*12/cleaned_farm_data["Pastureland Acres"]
```

```python
cleaned_farm_data["Chicken CO2 per Acre (Kg/Year)"] = cleaned_farm_data['Broilers and other meat-type chickens']*6*2.6*.62/cleaned_farm_data["Pastureland Acres"]
cleaned_farm_data["Cow CO2 per Acre (Kg/Year)"] = cleaned_farm_data['Cattle and calves']*60*340/cleaned_farm_data["Pastureland Acres"]
cleaned_farm_data["Sheep CO2 per Acre (Kg/Year)"] = cleaned_farm_data['Sheep and lambs']*20*30/cleaned_farm_data["Pastureland Acres"]
cleaned_farm_data["Pig CO2 per Acre (Kg/Year)"] = cleaned_farm_data['Hogs and pigs']*7*130*.7/cleaned_farm_data["Pastureland Acres"]
```

```python
cleaned_farm_data["Chicken N2O per Acre (Kg/Year)"] = cleaned_farm_data['Broilers and other meat-type chickens']*.00004*365/cleaned_farm_data["Pastureland Acres"]
cleaned_farm_data["Cow N2O per Acre (Kg/Year)"] = cleaned_farm_data['Cattle and calves']*.48*365/cleaned_farm_data["Pastureland Acres"]
cleaned_farm_data["Sheep N2O per Acre (Kg/Year)"] = cleaned_farm_data['Sheep and lambs']*.16*365/cleaned_farm_data["Pastureland Acres"]
cleaned_farm_data["Pig N2O per Acre (Kg/Year)"] = cleaned_farm_data['Hogs and pigs']*.83*365/cleaned_farm_data["Pastureland Acres"]
```
